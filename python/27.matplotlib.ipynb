```python
x = [1, 2, 3, 4, 5]
y1 = [5, 1, 7, 8, 2]
y2 = [7, 8, 3, 5, 1]

plt.plot(x, y1, color='red', marker='o', markersize=6, linestyle='--', linewidth=3, label="line y1 as tata motors")
plt.plot(x, y2, color='b', marker='o', markersize=6, linestyle='--', linewidth=3, label="line y2 as tata power")

plt.title("Multiple line in one plot")

plt.xlabel("Days")
plt.ylabel("stack price")
plt.legend()

plt.show()
```

```python
#pie char >> To understand the percent distribution of data
langs = ["C", "Python", "Java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = "%1.1f%%")
plt.title("Pie chart")
plt.show()
```

```python
#focus on one segment
explode = (0.0, 0.1, 0.1)
langs = ["C", "Python", "Java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = "%1.1f%%", explode = explode)
plt.title("Pie chart")
plt.show()
```

```python
explode = (0.0, 0.1, 0.1)
langs = ["C", "Python", "Java"]
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = "%1.1f%%", explode = explode, shadow = True)
plt.title("Pie chart")
plt.show()
```

```python
years = [2021, 2018, 2022, 2023, 2017]
category1 = [20, 25, 30, 28, 35]
category2 = [10, 15, 20, 18, 25]
category3 = [5, 8, 10, 9, 15]

plt.stackplot(years, category1, category2, category3, labels = ["category1", "category2", "category3",],
             colors = ["skyblue","lightgreen", "orange"])
plt.xlabel("years")
plt.ylabel("Values")
plt.title("Area plot")
plt.legend()
plt.show()
```
