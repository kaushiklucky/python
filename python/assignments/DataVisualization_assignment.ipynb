```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
np.random.seed(42)
size = 1000
feature_1 = np.random.normal(0, 1, size)
feature_2 = 0.8 * feature_1 + np.random.normal(0, 0.2, size)
feature_3 = 0.6 * feature_1 - 0.3 * feature_2 + np.random.normal(0, 0.3, size)

data = pd.DataFrame({'Feature 1': feature_1, 'Feature 2': feature_2, 'Feature 3': feature_3})

corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 2. Heatmap for Sales Data

np.random.seed(20)

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

heatmap_data = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum')

fig = px.imshow(heatmap_data, labels={'color': 'Sales'}, x=heatmap_data.columns, y=heatmap_data.index)

fig.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(x + y)

import plotly.graph_objs as go

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
                  scene=dict(
                      xaxis_title='X Axis',
                      yaxis_title='Y Axis',
                      zaxis_title='Z Axis'),
                  width=800, height=800)

fig.show()
```

```python
# 5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.
import plotly.express as px

np.random.seed(25)

data = {
    'Country': ['Rassia', 'Japan', 'INDIA', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

fig = px.scatter(df, x='GDP', y='Population', size='Population', color='Country', 
                 hover_name='Country', size_max=60)

fig.update_layout(title='Bubble Chart: Population vs GDP',
                  xaxis_title='GDP (in billions)',
                  yaxis_title='Population (in millions)')

fig.show()
```

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

p.line(x, y, legend_label="Sine Wave", line_width=2, color='blue')

show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

output_notebook()

np.random.seed(0)
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10

sizes = np.random.rand(100) * 20
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=100)

data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

p = figure(title="Scatter Plot with Varying Sizes and Colors", x_axis_label='x', y_axis_label='y', width=800, height=400)

p.scatter(x='x', y='y', size='sizes', color='colors', fill_alpha=0.6, source=data)

show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', width=800, height=400)

p.vbar(x=fruits, top=counts, width=0.4, color="navy", alpha=0.7)

show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency', width=800, height=400)

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palettes import Viridis256
from bokeh.layouts import column

output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

p = figure(title="Heatmap", x_axis_label='x', y_axis_label='y', width=600, height=600,
           tools='hover,save,pan,box_zoom,wheel_zoom,reset', x_range=(0, 1), y_range=(0, 1))

mapper = linear_cmap(field_name='z', palette=Viridis256, low=min(z_flat), high=max(z_flat))

p.rect(x='x', y='y', width=0.1, height=0.1, source=dict(x=x_flat, y=y_flat, z=z_flat),
       line_color=None, fill_color=mapper, fill_alpha=0.8)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

show(p)
```
